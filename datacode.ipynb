```python
import pandas as pd

# Load CSV safely
df_csv = pd.read_csv("/Users/roamsaleh/AlBayanFi/arabic_poetry_cleaned.csv", 
                     encoding="utf-8", 
                     engine="python")

# Save as JSON
csv_as_json_path = "/Users/roamsaleh/AlBayanFi/arabic_poetry_cleaned.json"
df_csv.to_json(csv_as_json_path, orient="records", force_ascii=False, indent=2)

print("CSV converted to JSON:", csv_as_json_path)
```

```python
import json

path1 = "/Users/roamsaleh/AlBayanFi/dsaa.json"
path2 = "/Users/roamsaleh/AlBayanFi/data.json"
output_path = "/Users/roamsaleh/AlBayanFi/merged_result.json"
with open(path1, 'r', encoding='utf-8') as f1:
    json1_data = json.load(f1)

with open(path2, 'r', encoding='utf-8') as f2:
    json2_data = json.load(f2)

merged_by_id = {}

for item in json1_data:
    item_id = item.get('poem_id') 
    if item_id:
        merged_by_id[item_id] = item

for item in json2_data:
    item_id = item.get('poem_id')
    if item_id in merged_by_id:
        merged_by_id[item_id].update(item)
    else:
        merged_by_id[item_id] = item

merged_list = list(merged_by_id.values())
with open(output_path, 'w', encoding='utf-8') as f_out:
    json.dump(merged_list, f_out, indent=4, ensure_ascii=False)

print(f"ok {output_path}")
```
